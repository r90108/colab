```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

1.讀取資料:

df= pd.read_csv('data.csv') #讀取csv檔案

2.檢視資料:

df.head() 查看前五筆資料

df.tail() 查看最後五筆資料

df.shape 查看資料框的形狀

df.columns 查看所有欄位名稱

df.describe() 查看統計量，如平均值、中位數、標準差等

df.info() 查看資料框的資訊

3.篩選資料:

df['column_name'] 選取某個欄位

df.loc[row_indexer, column_indexer] 使用行索引器和列索引器進行資料選擇

df.iloc[row_indexer, column_indexer] 使用整數索引器進行資料選擇

df[(df['column_name'] == 'value')] 條件篩選資料

4.排序資料:

df.sort_values(by='column_name') 按照欄位名稱進行升序排序

df.sort_values(by='column_name', ascending=False) 按照欄位名稱進行降序排序

df.sort_values(by=['column_name_1', 'column_name_2']) 按照多個欄位進行排序

5.處理遺漏值:

df.dropna() 刪除所有包含遺漏值的列

df.fillna(value) 用value填補所有遺漏值

df.isna() 檢查哪些值是遺漏值

6.處理重複值:

df.drop_duplicates() 刪除所有重複的行

7.計算統計量:

df.mean() 計算所有欄位的平均值

df.std() 計算所有欄位的標準差

df.max() 計算所有欄位的最大值

df.min() 計算所有欄位的最小值

8.資料的聚合和分組:

df.groupby('column_name').sum() 將資料按照某個欄位進行分組，並計算每組的總和

df.groupby(['column_name_1', 'column_name_2']).mean() 將資料按照多個欄位進行分組，並計算每組的平均值

9.自訂dataframe:

新dataframe =pd.DataFrame(array, columns = 項目list)
